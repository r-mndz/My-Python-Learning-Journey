Identify if it's a prime number

```python

#---> this program identify wether the user input a prime number or not.
def is_prime(n):
    if n <= 1:
      return False
    if n == 2:
      return True
    if n % 2 == 0:
      return False
    for i in range(3, int(n**0.5) + 1, 2):
      if n % i == 0:
        return False
    return True
n = int(input("Enter a number: "))
print(f"{n} is a prime:", is_prime(n))
```

```python
# ---> this program cleaned user input remove spaces and capitalize properly
# ---> .strip() remove white space
# --->  .append(), to connect the new list to a new declared list
# --->  .title() to properly capitalized list of stripe
# --->  .join() to join list
SHOWS = [
    " Avatar: the last airbender",
    "Ben 10",
    "Athur",
    " Spongebob Squarepants",
    "Phineas and ferb",
    "Kim possible",
    "Jimmy Neutron ",
    "the Proud family"
]

def main():
    cleaned_shows = []   # ---> here I created a new list to of cleaned data
    for show in SHOWS:
        cleaned_shows.append((show.strip().title()))  # ---> here I use the .append method to capture the cleaned data
    print(', '.join(cleaned_shows))  #---> here is where I print the new list from cleaned data.

main()
```
